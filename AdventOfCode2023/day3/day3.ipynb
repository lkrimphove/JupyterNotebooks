```python
INPUT = 'input.txt'

file = open(INPUT, 'r')
lines = file.readlines()
```

```python
# PART ONE

def is_symbol(char):
    if char.isdigit() or char == '.':
        return False
    else:
        return True


schematic = []
for line in lines:
    schematic.append(list(line.strip()))

numbers = []
for y in range(len(schematic)):
    for x in range(len(schematic[y])):
        char = schematic[y][x]
        if is_symbol(char):
            # check all around symbol for a digit
            for j in range(-1,2):
                for i in range (-1,2):
                    # look at each adjacent character
                    if y + j < 0 or y + j >= len(schematic):
                        continue
                    if x + i < 0 or x + i >= len(schematic[y + j]):
                        continue

                    if schematic[y + j][x + i].isdigit():
                        # found a digit, but need to get the whole number
                        number = ''

                        left_pointer = x + i
                        left_value = schematic[y + j][left_pointer]
                        while left_value.isdigit():
                            number = left_value + number
                            # delete current value from schematic to prevent duplicate numbers
                            schematic[y + j][left_pointer] = '.'

                            left_pointer -= 1
                            if left_pointer >= 0:
                                left_value = schematic[y + j][left_pointer]
                            else:
                                left_value = '.'

                        
                        right_pointer = x + i + 1
                        right_value = schematic[y + j][right_pointer]
                        while right_value.isdigit():
                            number = number + right_value

                            # delete current value from schematic to prevent duplicate numbers
                            schematic[y + j][right_pointer] = '.'
                            
                            right_pointer += 1
                            if right_pointer < len(schematic[y + j]):
                                right_value = schematic[y + j][right_pointer]
                            else:
                                right_value = '.'
                        
                        numbers.append(int(number))

print(sum(numbers))
```

```python
# PART TWO

def is_gear(char):
    if char == '*':
        return True
    else:
        return False


schematic = []
for line in lines:
    schematic.append(list(line.strip()))

ratios = []
for y in range(len(schematic)):
    for x in range(len(schematic[y])):
        char = schematic[y][x]
        if is_gear(char):
            numbers = []
            # check all around symbol for a digit
            for j in range(-1,2):
                for i in range (-1,2):
                    # look at each adjacent character
                    if y + j < 0 or y + j >= len(schematic):
                        continue
                    if x + i < 0 or x + i >= len(schematic[y + j]):
                        continue

                    if schematic[y + j][x + i].isdigit():
                        # found a digit, but need to get the whole number
                        number = ''

                        left_pointer = x + i
                        left_value = schematic[y + j][left_pointer]
                        while left_value.isdigit():
                            number = left_value + number
                            # delete current value from schematic to prevent duplicate numbers
                            schematic[y + j][left_pointer] = '.'

                            left_pointer -= 1
                            if left_pointer >= 0:
                                left_value = schematic[y + j][left_pointer]
                            else:
                                left_value = '.'

                        
                        right_pointer = x + i + 1
                        right_value = schematic[y + j][right_pointer]
                        while right_value.isdigit():
                            number = number + right_value

                            # delete current value from schematic to prevent duplicate numbers
                            schematic[y + j][right_pointer] = '.'
                            
                            right_pointer += 1
                            if right_pointer < len(schematic[y + j]):
                                right_value = schematic[y + j][right_pointer]
                            else:
                                right_value = '.'
                        
                        numbers.append(int(number))

            if len(numbers) == 2:
                ratios.append(numbers[0] * numbers[1])

print(sum(ratios))
```
